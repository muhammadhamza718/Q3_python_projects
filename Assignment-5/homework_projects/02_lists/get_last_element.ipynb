```python
def get_last_element(lst):
    """
    Prints the last element of the provided list.
    """
    print(lst[-1])  # Access the last element directly using negative indexing

def get_lst():
    """
    Prompts the user to enter elements of the list one at a time and returns the list.
    """
    lst = []
    elem = input("Please enter an element of the list or press enter to stop: ")
    while elem != "":
        lst.append(elem)
        elem = input("Please enter an element of the list or press enter to stop: ")
    return lst

def main():
    lst = get_lst()  # Get user-inputted list
    get_last_element(lst)  # Print the last element

# Run the function
main()
```
